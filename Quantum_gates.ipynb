# Pauli X gate implementation

```python
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator

# 1. Build the Abstract Circuit
# Define intent using virtual qubits without baking in physical constraints yet
abstract_circuit = QuantumCircuit(1, 1)
abstract_circuit.x(0)         # Apply Pauli-X gate
abstract_circuit.measure(0, 0) # Measure qubit 0 to classical bit 0

# 2. Establish the Backend Target
# In modern Qiskit workflows, a backend acts as a structural reference target
backend = AerSimulator()

# 3. IBM Norm: Compile to an ISA (Instruction Set Architecture) Physical Circuit
# 2026 standards require explicit transpilation targeting your backend before execution
pass_manager = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pass_manager.run(abstract_circuit)

# 4. Execute the Compiled Physical Circuit
job = backend.run(isa_circuit, shots=1024)
result = job.result()

# 5. Extract Data and Render Diagram
counts = result.get_counts()
print("Execution Results:", counts)

# Draw the compliant physical circuit layout
isa_circuit.draw('mpl')
```

# Phase Shift gate

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator

# 1. Build the Abstract Circuit
abstract_circuit = QuantumCircuit(1, 1)
abstract_circuit.h(0)                 # Put qubit into superposition
abstract_circuit.p(np.pi / 2, 0)      # Apply Phase Shift gate with theta = pi/2
abstract_circuit.measure(0, 0)         # Measure qubit 0 into classical bit 0

# 2. Establish the Backend Target
backend = AerSimulator()

# 3. IBM Norm: Explicitly compile abstract intent to an ISA Physical Circuit
pass_manager = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pass_manager.run(abstract_circuit)

# 4. Execute the Compiled Physical Circuit
job = backend.run(isa_circuit, shots=1024)
result = job.result()

# 5. Extract Data and Render Diagram
counts = result.get_counts()
print("Execution Results:", counts)

# Draw the compliant physical circuit layout
isa_circuit.draw('mpl')
```

# 2 qubit X gate

```python
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator

# 1. Build the Abstract Circuit
# We define 2 qubits and 2 classical bits
abstract_circuit = QuantumCircuit(2, 2)

# Apply Pauli-X gates to both qubit 0 and qubit 1
abstract_circuit.x(0)
abstract_circuit.x(1)

# Measure both qubits into their corresponding classical bits
# [0, 1] specifies the qubits, [0, 1] specifies the destination classical bits
abstract_circuit.measure([0, 1], [0, 1])

# 2. Establish the Backend Target
backend = AerSimulator()

# 3. IBM Norm: Compile abstract intent to an ISA Physical Circuit
pass_manager = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pass_manager.run(abstract_circuit)

# 4. Execute the Compiled Physical Circuit
job = backend.run(isa_circuit, shots=1024)
result = job.result()

# 5. Extract Data and Render Diagram
counts = result.get_counts()
print("Execution Results:", counts)

# Draw the compliant physical circuit layout
isa_circuit.draw('mpl')
```

# Swap Gate

```python
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator

# 1. Build the Abstract Circuit
abstract_circuit = QuantumCircuit(2, 2)

# Set up an initial state: Flip qubit 0 to |1> (qubit 1 stays |0>)
abstract_circuit.x(0)

# Apply the SWAP gate to exchange the states of qubit 0 and qubit 1
abstract_circuit.swap(0, 1)

# Measure both qubits into their corresponding classical bits
abstract_circuit.measure([0, 1], [0, 1])

# 2. Establish the Backend Target
backend = AerSimulator()

# 3. IBM Norm: Compile abstract intent to an ISA Physical Circuit
pass_manager = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pass_manager.run(abstract_circuit)

# 4. Execute the Compiled Physical Circuit
job = backend.run(isa_circuit, shots=1024)
result = job.result()

# 5. Extract Data and Render Diagram
counts = result.get_counts()
print("Execution Results:", counts)

# Draw the compliant physical circuit layout
isa_circuit.draw('mpl')
```

# Toffli gate(3 qubit)

```python
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator

# 1. Build the Abstract Circuit
# We define 3 qubits (2 controls, 1 target) and 3 classical bits
abstract_circuit = QuantumCircuit(3, 3)

# Prepare initial states: Flip both control qubits (0 and 1) to |1>
abstract_circuit.x(0)
abstract_circuit.x(1)

# Apply the Toffoli (CCX) gate
# Arguments: control_1, control_2, target
abstract_circuit.ccx(0, 1, 2)

# Measure all qubits into their corresponding classical bits
abstract_circuit.measure([0, 1, 2], [0, 1, 2])

# 2. Establish the Backend Target
backend = AerSimulator()

# 3. IBM Norm: Compile abstract intent to an ISA Physical Circuit
pass_manager = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pass_manager.run(abstract_circuit)

# 4. Execute the Compiled Physical Circuit
job = backend.run(isa_circuit, shots=1024)
result = job.result()

# 5. Extract Data and Render Diagram
counts = result.get_counts()
print("Execution Results:", counts)

# Draw the compliant physical circuit layout
isa_circuit.draw('mpl')
```

# Fredkin Gate(3 qubit)

```python
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_aer import AerSimulator

# 1. Build the Abstract Circuit
# We define 3 qubits and 3 classical bits
abstract_circuit = QuantumCircuit(3, 3)

# Prepare initial states:
abstract_circuit.x(0)  # Set Control (qubit 0) to |1>
abstract_circuit.x(1)  # Set Target 1 (qubit 1) to |1> (Target 2/qubit 2 stays |0>)

# Apply the Fredkin (Controlled-SWAP) gate
# Arguments: control, target_1, target_2
abstract_circuit.cswap(0, 1, 2)

# Measure all qubits into their corresponding classical bits
abstract_circuit.measure([0, 1, 2], [0, 1, 2])


# 2. Establish the Backend Target
backend = AerSimulator()

# 3. IBM Norm: Compile abstract intent to an ISA Physical Circuit
pass_manager = generate_preset_pass_manager(optimization_level=1, backend=backend)
isa_circuit = pass_manager.run(abstract_circuit)

# 4. Execute the Compiled Physical Circuit
job = backend.run(isa_circuit, shots=1024)
result = job.result()

# 5. Extract Data and Render Diagram
counts = result.get_counts()
print("Execution Results:", counts)

# Draw the compliant physical circuit layout
isa_circuit.draw('mpl')
```
